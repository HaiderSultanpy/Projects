```python
import requests
import pandas as pd
import io  # Import the io module

# URL of the CSV file
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Send a HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Load the content of the response into a DataFrame
    df = pd.read_csv(io.StringIO(response.text))
    print(df.head())  # Display the first few rows of the DataFrame
else:
    print(f"Failed to fetch the file. Status code: {response.status_code}")
```

```python
#create data for plotting
df_line = df.groupby(df['Year'])['Automobile_Sales'].mean()
    #create figure
plt.figure(figsize=(10, 6))
df_line.plot(kind = 'line')
plt.xlabel('........')
plt.ylabel('.........')
plt.title('......................')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_line = df.groupby(df['Year'])['Automobile_Sales'].mean()
df_line.plot(kind = 'line')
plt.xticks(list(range(1980,2024)), rotation = 90)
plt.xlabel('year')
plt.ylabel('sales')
plt.title('Automobile sales in Recession')
plt.text(1982, 650, '1981-82 Recession')
plt.text(1991,600 , ' 1991 Recession')
plt.text(2001,600 , ' 2000-2001 Recession')
plt.text(2007,500 , ' 2007-2009 Recession')
plt.text(2020,2000 , ' COVID Recession')
plt.legend()
plt.show()
```

```python
df_rec = df[df['Recession']==1]
df_Mline = df_rec.groupby(['Year','Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()
df_Mline.set_index('Year', inplace=True)
df_Mline = df_Mline.groupby(['Vehicle_Type'])['Automobile_Sales']
df_Mline.plot(kind='line')
plt.xlabel('..............')
plt.ylabel('............')
plt.title('Sales Trend Vehicle-wise during Recession')
plt.legend()
plt.show()
```

```python
#To visualize the average number of vehicles sold during recession and non-recession periods
new_df = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()

# Create the bar chart using seaborn
plt.figure(figsize=(4, 5))
sns.barplot(x='Recession', y='Automobile_Sales', hue='Recession',  data=new_df)
plt.xlabel('............')
plt.ylabel('...............')
plt.title('Average Automobile Sales during Recession and Non-Recession')
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.show()
```

```python
#To visualize sales of different vehicles during recession and non-recession periods
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the data for recessionary periods
recession_data = df[df['Recession'] == 1]

# Calculate the average automobile sales by recession and vehicle type
dd = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Create the grouped bar chart using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Recession', y='Automobile_Sales', hue='Vehicle_Type', data=dd)
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.xlabel('Economic Period')
plt.ylabel('Average Automobile Sales')
plt.title('Vehicle-Wise Sales during Recession and Non-Recession Periods')

plt.show()
```

```python
#How did the GDP vary over time during recession and non-recession periods?
import matplotlib.pyplot as plt

# Filter the data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Create a figure
fig = plt.figure(figsize=(14, 6))

# Add the first subplot for recession period
ax1 = fig.add_subplot(1, 2, 1)  # 1 row, 2 columns, position 1
ax1.plot(recession_data['Year'], recession_data['GDP'], marker='o', color='red')
ax1.set_title('GDP during Recession Period')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.grid(True)

# Add the second subplot for non-recession period
ax2 = fig.add_subplot(1, 2, 2)  # 1 row, 2 columns, position 2
ax2.plot(non_recession_data['Year'], non_recession_data['GDP'], marker='o', color='green')
ax2.set_title('GDP during Non-Recession Period')
ax2.set_xlabel('Year')
ax2.grid(True)

# Set the overall title for the figure
fig.suptitle('GDP Variations during Recession and Non-Recession Periods', fontsize=16)

# Adjust layout to prevent overlapping
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the data for recession and non-recession periods
Rdata = df[df['Recession'] == 1]
NRdata = df[df['Recession'] == 0]

# Calculate the total advertising expenditure for both periods
RAtotal = Rdata['Advertising_Expenditure'].sum()
NRAtotal = NRdata['Advertising_Expenditure'].sum()

# Create a pie chart for the advertising expenditure
plt.figure(figsize=(8, 6))

labels = ['Recession', 'Non-Recession']
sizes = [RAtotal, NRAtotal]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue'])

# Title of the pie chart
plt.title('Advertising Expenditure during Recession and Non-Recession Periods')

# Display the pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the data for recession periods
Rdata = df[df['Recession'] == 1]

# Calculate the total advertisement expenditure by vehicle type during recessions
VTexpenditure = Rdata.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Create a pie chart for the share of each vehicle type in total expenditure during recessions
plt.figure(figsize=(10, 8))

labels = VTexpenditure.index
sizes = VTexpenditure.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))

# Title of the pie chart
plt.title('Advertisement Expenditure by Vehicle Type during Recession Period')

# Display the pie chart
plt.show()
```

```python
import pandas as pd
import folium
import json

# Load the sales data (replace this with your actual DataFrame)
# df = pd.read_csv("path_to_your_sales_data.csv")  # Example: load your sales data

# Assuming the sales DataFrame has these columns: 'State', 'Automobile_Sales', 'Recession'
df_rec = df[df['Recession'] == 1]  # Filter for recession data

# Group by state and sum sales
state_sales = df_rec.groupby('State')['Automobile_Sales'].sum().reset_index()

# Load the US states GeoJSON data
with open("us-states.json") as f:
    states_geo = json.load(f)

# Create a map centered around the US
m = folium.Map(location=[37.8, -96], zoom_start=4)

# Add sales data to the map
for _, row in state_sales.iterrows():
    # Create a color based on sales value (for example)
    color = 'blue' if row['Automobile_Sales'] > 5000 else 'orange'  # You can customize the logic

    # Create a GeoJson layer for each state
    folium.GeoJson(
        states_geo,
        style_function=lambda x, color=color: {
            'fillColor': color,
            'color': 'black',
            'weight': 0.5,
            'fillOpacity': 0.7
        },
        tooltip=f"{row['State']}: {row['Automobile_Sales']}"
    ).add_to(m)

# Save the map to an HTML file
m.save('highest_sales_regions.html')

# To display the map in a Jupyter Notebook (if needed)
# display(m)
```
